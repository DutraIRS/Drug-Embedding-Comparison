```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\quiia\\Desktop\\Drug-Embedding-Comparison\\data\\R.csv", header=None, sep=';')
```

```python
import utils, torch
```

```python
smiles = list(df.iloc[:, 0].values)
side_effects_data = torch.tensor(df.iloc[:, 1:].values, dtype=torch.float)

full_dataset = utils.DrugSideEffectsDataset(smiles, side_effects_data)
```

```python
len(full_dataset.data)
```

```python
import seaborn as sns

sns.histplot([x[0].size()[0] for x in full_dataset.data])
```

```python
len([x[0].size()[0] for x in full_dataset.data if x[0].size()[0] <= 100])
```

```python
ind = []

for i, x in enumerate(full_dataset.data):
    if x[0].size()[0] <= 100:
        ind.append(i)

ind
```

```python
df = pd.read_csv("C:\\Users\\quiia\\Desktop\\Drug-Embedding-Comparison\\data\\R.csv", header=None, sep=';')

df = df.iloc[ind, :].reset_index(drop=True)

df.to_csv("C:\\Users\\quiia\\Desktop\\Drug-Embedding-Comparison\\data\\R_100.csv", header=None, sep=';', index=False)
```
